# Problem plecakowy

Optymalne rozwiązanie 1458

```python
def P(X,Wartości):
    return sum([a*b for a,b in zip(X,Wartości)])

def W(X,Wagi):
    return sum([a*b for a,b in zip(X,Wagi)])

def stan_plecaka(X,Przedmioty):
    plecak=[]
    for i in range(len(X)):
        if X[i]:
            plecak.append(Przedmioty[i])
    return plecak
```

```python
Przedmioty=['Woda','Chleb','Pochodnia','Szynka','Łuk','Lina','Talerz','Kubek','Termos','Nóż','Latarka','Kij','Buty','Miecz','Włócznia']
```

```python
Wartości=[135,139,149,150,156,163,173,184,192,201,210,214,221,229,240]
```

```python
Wagi=[70,73,77,80,82,87,90,94,98,106,110,113,115,118,120]
```

```python
Maxymalna_waga=750
```

Symulowane wyżarzanie

```python
best=[]
```

```python
def symulowane_wyżarzanie(Wartości,Wagi,M,cmax,T0,a):
    c=0
    T=T0
    X=[0]*len(Wagi)
    Aktualna_waga=0
    Xbest=X.copy()
    while(c<=cmax):
        j=randrange(len(Wagi))
        Y=X.copy()
        Y[j]=1-X[j]
        if(Y[j]==1 and Aktualna_waga+Wagi[j]>M):
            Y=None
        if(Y != None):
            if(Y[j]==1):
                X=Y.copy()
                Aktualna_waga=Aktualna_waga+Wagi[j]
                if(P(X,Wartości)>P(Xbest,Wartości)):
                    Xbest=X.copy()
            else:
                r=random.uniform(0, 1)
                if(r<math.exp(-(Wartości[j]/T))):
                    X=Y.copy()
                    Aktualna_waga=Aktualna_waga-Wagi[j]
        c=c+1
        T=T*a
        
        print(str(c)+'\n'+str(stan_plecaka(X,Przedmioty))+'\n'+str(P(X,Wartości)))
        print('_______________________________________')
        best.append(P(X,Wartości))
        
    print('______________________________________________________________________')
    print(stan_plecaka(Xbest,Przedmioty))
    return P(Xbest,Wartości)
        
```

```python
symulowane_wyżarzanie(Wartości,Wagi,Maxymalna_waga,cmax=100,T0=10000,a=0.9999)
```

```python
plt.plot(best[:])
```

```python
for c in [100,1000,5000,10000]:
    lst1.append(symulowane_wyżarzanie(Wartości,Wagi,Maxymalna_waga,cmax=c,T0=10000,a=0.9999))
```

Przeszukiwanie tabu

```python
best=[]
```

```python
def find_max_i(N,X,Wagi,Wartości):
    a=-99999999
    b=-1
    for i in N:
        if math.pow(-1,X[i])*Wartości[i]/Wagi[i]>a:
            a=math.pow(-1,X[i])*Wartości[i]/Wagi[i]
            b=i
    return b
```

```python
def przeszukiwanie_tabu(Wartości,Wagi,Maxymalna_waga,cmax=100,L=1):
    c=1
    X=[0]*len(Wagi)
    Aktualna_waga=0
    Xbest=X.copy()
    tabu=[]
    while(c<=cmax):
        N=[i for i in range(len(Wagi))]
        start=max(0,c-L)
        for j in range(start,c-1):
            N.remove(tabu[j])
        for i in N.copy():
            if(X[i]==0 and (Aktualna_waga+Wagi[i]>Maxymalna_waga)):
                N.remove(i)
        print(N)
        if not N:
            break;
        i=find_max_i(N,X,Wagi,Wartości)
        tabu.append(i)
        X[i]=1-X[i]
        if(X[i]==1):
            Aktualna_waga=Aktualna_waga+Wagi[i]
        else:
            Aktualna_waga=Aktualna_waga-Wagi[i]
        if(P(X,Wartości)>P(Xbest,Wartości)):
            Xbest=X.copy()
        c=c+1
        print(str(c)+'\n'+str(stan_plecaka(X,Przedmioty))+'\n'+str(P(X,Wartości)))
        print('_______________________________________')
        best.append(P(X,Wartości))
        
    print('______________________________________________________________________')
    print(stan_plecaka(Xbest,Przedmioty))
    return P(Xbest,Wartości)
```

```python
przeszukiwanie_tabu(Wartości,Wagi,Maxymalna_waga,cmax=1000,L=15)
```
